```python
import pandas as pd
%matplotlib inline
```

```python
filename = "K:/Users/redddog/AIT/data/ait_data/n-alkanes/n-docosane/n-docosane_200415_1200_100_225.csv"
header_row = 0
with open(filename, 'r') as f:
    for i, line in enumerate(f.readlines()):
        if "time,t1,t2,t3,t4,pressure" in line:
            header_row = i
            break

df = pd.read_csv(filename, header=header_row, index_col=False, usecols=["time","t1","t2","t3","t4","pressure",])
df.head()
```

```python
ax=None
for t in ['t1', 't2', 't3', 't4']:
    ax = df.plot('time', t, ax=ax)
```
