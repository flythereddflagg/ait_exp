```python
import pandas as pd
%matplotlib inline
```

```python
filename = "/home/aitra/Documents/data/n-hexacosane_mod_altitude/n-hexacosane_mod_alt_200831_1536_100_331.csv"
header_row = 0
with open(filename, 'r') as f:
    for i, line in enumerate(f.readlines()):
        print("line", line)
        if "time,t1,t2,t3,t4,pressure" in line:
            header_row = i
            break

print(header_row)
df = pd.read_csv(
    filename,
    index_col=False, 
    usecols=["time","t1","t2","t3","t4","pressure"],
    skiprows=2
)
df.head()
```

```python
ax=None
for t in ['t1', 't2', 't3', 't4', 'pressure']:
    ax = df.plot('time', t, ax=ax)
```
